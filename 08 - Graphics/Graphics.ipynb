# กราฟิกส์

Swing มีเครื่องมือที่ทำให้เราสามารถสร้างภาพกราฟิกส์ต่าง ๆ ได้ง่าย ๆ เราสามารถวาดเส้น วาดรูปร่างต่าง ๆ และโหลดรูปขึ้นมาจัดวางได้ตามต้องการ รวมไปถึงสามารถตอบสนองต่อเมาส์และแป้นพิมพ์ได้ ทำให้เราสามารถสร้างเกมหรือโปรแกรมกราฟิกส์ที่มีการตอบโต้กับผู้ใช้ได้

การวาดบนโปรแกรมของเรา เราจะใช้ `JPanel` เป็นเหมือนผืนผ้าใบ (canvas) ให้เราวาดลงไปได้ โดยมีอ็อบเจกต์ `Graphics` เป็นเหมือนปากกาหรือพู่กันสำหรับวาด

## กลไกการทำงานของกราฟิกส์บน Swing

คอมโพเนนต์ที่เป็นคอนเทนเนอร์ใน Swing สามารถแบ่งได้เป็นคอนเทนเนอร์ชั้นบนสุด (top-level container) ได้แก่ `JFrame`, `JDialog` และ `JApplet` กับคอนเทนเนอร์ชั้นในซึ่งได้แก่คอมโพเนนต์ของ Swing ตัวอื่นที่เหลือ เช่น `JPanel` เป็นต้น

คอนเทนเนอร์ชั้นบนสุดจะเป็นเหมือนหน้าต่าง ซึ่งเป็นกรอบที่อยู่ชั้นนอกสุด และในหน้าต่างนั้นจะบรรจุคอมโพเนนต์อื่น ๆ เอาไว้

คอนเทนเนอร์ชั้นในและคอมโพเนนต์อื่น ๆ ที่ไม่ใช่คอนเทนเนอร์ชั้นบนสุด เช่น `JButton`, `JLabel` และคอมโพเนนต์อื่น ๆ จะมีเมทอด `paintComponent()` ซึ่งจะถูกเรียกทุกครั้งที่คอมโพเนนต์เหล่านั้นต้องมีการวาดตัวเองใหม่ เช่น เวลาที่เราเพิ่งเปิดหน้า GUI ขึ้นมา เวลาที่เราเปลี่ยนขนาดหน้าต่างโปรแกรม เวลาที่เราย้ายหน้าต่างโปรแกรม หรือเวลาที่เราย่อหน้าต่างลงไปแล้วเปิดกลับขึ้นมาใหม่

เหตุการณ์เหล่านั้นจะทำให้จำเป็นต้องมีการวาดหน้า GUI ใหม่ คอมโพเนนต์ชั้นในทุกตัวจะถูกเรียกเมทอด `paintComponent()` ให้เกิดการวาดตัวเองขึ้น

### คลาส JPanel

`JPanel` เป็นคลาสสำคัญที่เราใช้ในการสร้างกราฟิกส์บน Swing เพราะว่า `JPanel` เป็นคอมโพเนนต์ว่าง ๆ ที่เราสามารถวาดหรือวางอะไรลงไปก็ได้

การใช้คลาส `JPanel` ทำกราฟิกส์จะทำได้โดยการสร้างซับคลาสของ `JPanel` แล้วโอเวอร์ไรด์เมทอด `paintComponent()` ของ `JPanel` และสั่งวาดตามต้องการ

เมทอด `paintComponent()` จะถูกเรียกได้ใน 2 กรณี

1. ระบบพิจารณาแล้วเห็นว่าจำเป็นต้องมีการวาดหน้า GUI หรือคอมโพเนนต์บางตัวใหม่ ซึ่งเกิดขึ้นได้จากการเปิดหน้า GUI กลับขึ้นมา การเปลี่ยนขนาดหรือย้ายตำแหน่ง และถูกบดบัง เป็นต้น

2. โปรแกรมต้องการให้เกิดการวาดคอมโพเนนต์นั้นใหม่เองโดยการเรียกเมทอด `repaint()` ซึ่งจะทำให้เกิดการส่งคำร้องขอวาดใหม่ไปยัง EDT (event dispatch thread) การวาดใหม่จะเกิดกับคอมโพเนนต์นั้นและคอมโพเนนต์อื่นที่บรรจุอยู่ในคอมโพเนนต์นั้นอีกที

การสั่ง `repaint()` จะไม่ทำให้เกิดการวาดใหม่ทันที แต่จะเป็นการร้องขอให้วาดใหม่ การวาดใหม่จะเกิดขึ้นเมื่อ EDT พร้อมที่จะวาด โดยอาจจะรวบคำขอที่เกิดขึ้นหลังการวาดครั้งก่อนหน้าหลาย ๆ คำขอแล้ววาดใหม่พร้อมกันในคราวเดียวเลย

> **หมายเหตุ** อย่าเรียก `paintComponent()` โดยตรง ถ้าต้องการให้เกิดการวาดใหม่ให้เรียก `repaint()`

## คลาส Graphics2D

เมทอด `paintComponent()` รูปประกาศเต็มคือ `protected void paintComponent(Graphics g)` ซึ่งรับพารามิเตอร์ตัวเดียว มีชนิดเป็นคลาส `Graphics` ซึ่งเป็นคลาสนามธรรม เมื่อเมทอด `paintComponent()` ถูกเรียก อาร์กิวเมนต์ที่ถูกส่งเข้ามาคืออ็อบเจกต์จากคลาส `Graphics2D` ซึ่งเป็นซับคลาสของ `Graphics`

`Graphics` หรือ `Graphics2D` นี้คือ graphics context หรือตัวจัดการกราฟิกส์สำหรับคอมโพเนนต์นั้น ๆ เราสามารถมองคลาส `Graphics2D` นี้ได้เสมือนกับปากกา เราสามารถสั่งวาดเส้น รูปร่างต่าง ๆ ลงสี ได้โดยใช้อ็อบเจกต์ที่ถูกส่งเข้ามานี้ เช่น ในตัวอย่างนี้

````java
import java.awt.Graphics;
import javax.swing.JPanel;

public class DrawPanel extends JPanel {
    @Override
    public void paintComponent(Graphics g) {
        super.paintComponent(g);
        
        int width = getWidth();
        int height = getHeight();
        
        g.drawLine(0, 0, width, height);
        g.drawLine(0, height, width, 0);
    }
}
````

ซึ่งโอเวอร์ไรด์เมทอด `paintComponent()` ของ `JPanel` ด้วย การโอเวอร์ไรด์นี้ควรจะเรียก `paintComponent()` ของซูเปอร์คลาสด้วยเพื่อให้มีการวาดพื้นหลังตามปกติก่อนจะให้เราลงมือวาดเพิ่มเติม

ส่วนของหน้าต่างหลักเป็นดังนี้

````java
import javax.swing.JFrame;

public class DrawPanelTest {
    public static void main(String[] args) {
        DrawPanel panel = new DrawPanel();
        
        JFrame application = new JFrame();
        
        application.setDefaultCloseOperation(JFrame.EXIT_ON_CLOSE);
        
        application.add(panel);
        application.setSize(250, 250);
        application.setVisible(true);
    }
}
````

![DrawPanel02](DrawPanel02.png)

ในตัวอย่างนี้ เราเรียก `getWidth()` และ `getHeight()` ของ `JPanel` เพื่อให้รู้มิติของพื้นที่ที่วาดได้ของ `JPanel` นั้น และใช้เมทอด `drawLine()` ของคลาส `Graphics` เพื่อวาดเส้นลงบนพาเนล เมทอด `drawLine()` มีรูปแบบการใช้งานดังนี้

```java
void drawLine(int x1, int y1, int x2, int y2)
```

โดย `x1` และ `y1` แทนพิกัดของจุดตั้งต้น และ `x2` และ `y2` แทนพิกัดของจุดปลายเส้น

ระบบพิกัดของ Swing จะมีตำแหน่งบนซ้ายเป็นจุดกำเนิด มีพิกัดตามระบบพิกัด (x, y) เป็น (0, 0) และค่า x เพิ่มค่าไปทางขวา ค่าของ y เพิ่มลงไปทางด้านล่าง

นอกจาก `drawLine()` แล้ว `Graphics` ยังมีเมทอดอื่น ๆ ที่น่าสนใจ เช่น

- `setColor(Color c)` กำหนดสีให้เส้นหรือการลงสี
- `drawLine(int x1, int y1, int x2, int y2)` วาดส่วนของเส้นตรง
- `drawPolyline(int[] xPoints, int[] yPoints, int nPoints)` วาดส่วนของเส้นตรงต่อกัน อาร์เรย์ของพิกัดเป็นตัวระบุตำแหน่งของจุดเชื่อมต่อ `nPoints` จุด
- `drawRect(int x, int y, int width, int height)` วาดเส้นสี่เหลี่ยมโดยมีพิกัดบนซ้ายกำหนดโดย `x` และ `y` และขนาดกำหนดโดย `width` และ `height`
- `fillRect(int x, int y, int width, int height)` วาดสี่เหลี่ยมเต็มโดยมีพิกัดบนซ้ายกำหนดโดย `x` และ `y` และขนาดกำหนดโดย `width` และ `height`
- `drawOval(int x, int y, int width, int height)` วาดเส้นวงรีในกรอบสี่เหลี่ยมที่ระบุ
- `fillOval(int x, int y, int width, int height)` วาดวงรีเต็มในกรอบสี่เหลี่ยมที่ระบุ
- `drawArc(int x, int y, int width, int height, int startAngle, int arcAngle)` วาดเส้นส่วนโค้งในกรอบสี่เหลี่ยมที่ระบุ เรากำหนดส่วนโค้งโดยการระบุองศาเริ่มต้นและองศาสิ้นสุด
- `fillArc(int x, int y, int width, int height, int startAngle, int arcAngle)` วาดส่วนโค้งแบบเติมเต็มในกรอบสี่เหลี่ยมที่ระบุ เรากำหนดส่วนโค้งโดยการระบุองศาเริ่มต้นและองศาสิ้นสุด
- `drawPolygon(int[] xPoints, int[] yPoints, int nPoints)` วาดเส้นรูปหลายเหลี่ยม อาร์เรย์ของพิกัดเป็นตัวระบุตำแหน่งของมุมทั้ง `nPoints` มุม
- `fillPolygon(int[] xPoints, int[] yPoints, int nPoints)` วาดรูปหลายเหลี่ยมเต็ม อาร์เรย์ของพิกัดเป็นตัวระบุตำแหน่งของมุมทั้ง `nPoints` มุม
- `drawImage(Image img, int x, int y, ImageObserver observer)` วาดรูปภาพ พารามิเตอร์ `observer` มักจะกำหนดให้เป็น `null`
- `setFont(Font font)` กำหนดรูปแบบและขนาดของตัวอักษร
- `drawString(String str, int x, int y)` วาดข้อความตามฟอนต์และสีที่ตั้งไว้

## การจัดการเมาส์

เราสามารถรับข้อมูลจากเมาส์ได้โดยการจัดการกับอีเวนต์ที่เกี่ยวข้องกับเมาส์ ซึ่งจะประกอบด้วย `MouseEvent` และ `MouseWheelEvent` ในที่นี้เราจะสนใจเฉพาะ `MouseEvent` ซึ่งจะเกิดขึ้นเมื่อมีการคลิกเมาส์และขยับเมาส์ ส่วน `MouseWheelEvent` ซึ่งเกิดขึ้นเมื่อมีการหมุนล้อเลื่อน (scroll wheel) บนเมาส์เราไม่ได้นำมาพิจารณาในงานนี้

อินเทอร์เฟซ listener ที่เราใช้กับ `MouseEvent` จะมีอยู่ 2 แบบ ได้แก่ `MouseListener` ซึ่งจะจัดการกับเหตุการณ์การกดปุ่มเมาส์และการเลื่อนเมาส์เข้าหรือออกจากพื้นที่ของคอมโพเนนต์ และ `MouseMotionListener` ซึ่งจะจัดการกับการเลื่อนและลากตำแหน่งของเมาส์

### อินเทอร์เฟซ MouseListener

อินเทอร์เฟซ `MouseListener` มีไว้สำหรับจัดการการกดปุ่มเมาส์และการเลื่อนตำแหน่งเมาส์เข้าไปในพื้นที่ (enter) หรือเลื่อนออกมาจากพื้นที่ (exit) ของคอมโพเนนต์ ประกอบด้วยเมทอดดังนี้

- `void mouseClicked(MouseEvent e)` จะถูกเรียกเมื่อมีการคลิก (กดแล้วปล่อย) ปุ่มเมาส์
- `void mousePressed(MouseEvent e)` จะถูกเรียกเมื่อมีการกดปุ่มเมาส์
- `void mouseReleased(MouseEvent e)` จะถูกเรียกเมื่อมีการปล่อบปุ่มเมาส์
- `void mouseEntered(MouseEvent e)` จะถูกเรียกเมื่อตัวชี้ของเมาส์ข้ามเข้าไปในพื้นที่ของคอมโพเนนต์
- `void mouseExited(MouseEvent e)` จะถูกเรียกเมื่อตัวชี้ของเมาส์ข้ามออกมาจากพื้นที่ของคอมโพเนนต์

เมื่อเราอิมพลีเมนต์อินเทอร์เฟซ `MouseListener` เราต้องโอเวอร์ไรด์เมทอดทั้ง 5 เมทอดของ `MouseListener` เพื่อจัดการกับพฤติกรรมของเมาส์

### อินเทอร์เฟซ MouseMotionListener

อินเทอร์เฟซ `MouseMotionListener` มีไว้สำหรับการจัดการการขยับของเมาส์ ทั้งการเลื่อนเมาส์ปกติ และการคลิกปุ่มเมาส์ค้างไว้แล้วเลื่อน (เรียกว่าการลาก [drag]) ประกอบด้วยเมทอดดังนี้

- `void mouseMoved(MouseEvent e)` จะถูกเรียกเมื่อมีการขยับเมาส์
- `void mouseDragged(MouseEvent e)` จะถูกเรียกเมื่อมีการขยับเมาส์โดยที่มีการกดปุ่มค้างไว้ด้วย (การลาก)

ตัวอย่างต่อไปนี้เป็นการวาดด้วยเมาส์

````java
import java.awt.Graphics;
import java.awt.Color;
import java.awt.event.MouseEvent;
import java.awt.event.MouseListener;
import java.awt.event.MouseMotionListener;
import javax.swing.JPanel;
import java.util.ArrayList;

public class DrawingBoard extends JPanel {
    private ArrayList<Integer> xPoints = new ArrayList<>();
    private ArrayList<Integer> yPoints = new ArrayList<>();
    private ArrayList<Float> inkLevels = new ArrayList<>();

    private float inkLevel;

    private static final float INK_DEPLETION_RATE = 0.005f;
    private static final float INK_STOP_CONCENTRATION = 1.5f;
    private static final int INK_WIDTH = 12;

    public DrawingBoard() {
        super();

        setBackground(Color.WHITE);
        inkLevel = 0.0f;

        addMouseListener(new MouseListener() {
            @Override
            public void mouseClicked(MouseEvent e) {
                // Do nothing
            }

            @Override
            public void mousePressed(MouseEvent e) {
                xPoints.add(e.getX());
                yPoints.add(e.getY());
                inkLevel = 0.0f;
                inkLevels.add(inkLevel);
                repaint();
            }

            @Override
            public void mouseReleased(MouseEvent e) {
                xPoints.add(e.getX());
                yPoints.add(e.getY());
                inkLevel /= INK_STOP_CONCENTRATION;
                inkLevels.add(inkLevel);
                repaint();
            }

            @Override
            public void mouseEntered(MouseEvent e) {
                // Do nothing
            }

            @Override
            public void mouseExited(MouseEvent e) {
                // Do nothing
            }
        });

        addMouseMotionListener(new MouseMotionListener() {
            @Override
            public void mouseMoved(MouseEvent e) {
                // Do nothing
            }

            @Override
            public void mouseDragged(MouseEvent e) {
                xPoints.add(e.getX());
                yPoints.add(e.getY());
                inkLevel += INK_DEPLETION_RATE;
                if (inkLevel > 1.0f)
                    inkLevel = 1.0f;
                inkLevels.add(inkLevel);
                repaint();
            }
        });
    }

    @Override
    public void paintComponent(Graphics g) {
        super.paintComponent(g);

        for (int i = 0; i < xPoints.size(); i++) {
            float inkLevel = inkLevels.get(i);
            int width = (int)(INK_WIDTH * (0.2f + (1.0f - inkLevel) * 0.8f));
            g.setColor(new Color(inkLevel, inkLevel, inkLevel));
            g.fillOval(xPoints.get(i), yPoints.get(i), width, width);
        }
    }
}
````

คลาส `DrawingBoard` ซ้บคลาสจาก `JPanel` ทำหน้าที่เป็นกระดานเปล่าสำหรับวาดรูป โดยในส่วนของคอนสตรักเตอร์จะมีการกำหนดสีพื้นหลังให้เป็นสีขาว และกำหนดความเข้มของหมึกตั้งต้นให้เป็นสีดำสนิท (`0.0f`)

นอกจากนี้ยังมีการลงทะเบียนตัวจัดการอีเวนต์ไว้สำหรับตัวเองโดยใช้ทั้งอินเทอร์เฟซ `MouseListener` โดยโอเวอร์ไรด์เมทอด `MousePressed()` กับ `MouseReleased()` และอินเทอร์เฟซ `MouseMotionListener` โดยโอเวอร์ไรด์เมทอด `MouseDragged()`

เราให้การกดปุ่มเมาส์เป็นการเริ่มต้นการวาด โดยใน `MousePressed()` เราจะตั้งค่าความเข้มหมึกกลับไปเต็มใหม่ และบันทึกตำแหน่งและความเข้มหมึก ณ จุดเริ่มต้นลงไปใน `ArrayList` ที่เตรียมไว้

การค้างการคลิกเอาไว้แล้วลากเมาส์ถือเป็นการลากพู่กัน `MouseDragged()` จะถูกเรียก โดยเราจะลดความเข้มหมึกลงเรื่อย ๆ ตามค่า `INK_DEPLETION_RATE` ในแต่ละจุดที่เกิดอีเวนต์การลาก และบันทึกตำแหน่งกับความเข้ม ณ แต่ละจุดที่เกิดอีเวนต์เอาไว้ใน `ArrayList` เช่นกัน

เมื่อเราปล่อยปุ่มเมาส์ เราจะถือว่าเป็นการหยุดพู่กันและยกขึ้น `MouseReleased()` จะถูกเรียก ในจุดที่หยุดลากพู่กันเราจะให้มีการสะสมตัวของหมึก ทำให้จุดนั้นเข้มขึ้น เราเพิ่มความเข้มหมึกด้วยอัตรา `INK_STOP_CONCENTRATION` และบันทึกตำแหน่งกับความเข้ม ณ จุดนั้น

> **หมายเหตุ** สีดำสนิทมีความเข้มเป็น 0.0 และสีขาวสนิทมีความเข้มเป็น 1.0 การลดความเข้มของหมึกในที่นี้คือการเพิ่มค่าความเข้มของสีขาว

ทั้งนี้ การบันทึกตำแหน่งจะไม่เกิดผลใด ๆ เลยถ้าไม่เกิดการวาดขึ้นจริง ในเมทอดจัดการเมาส์เหล่านั้น จะเห็นว่าเรามีการเรียกเมทอด `repaint()` เพื่อแจ้ง EDT ว่าเราต้องการให้เกิดการวาด `DrawingBoard` นี้ (ซึ่งก็คือ `JPanel`) ขึ้นใหม่

เราโอเวอร์ไรด์เมทอด `paintComponent()` เพื่อให้เกิดการวาดขึ้น เมทอด `paintComponent()` จะเรียก `paintComponent()` ของซูเปอร์คลาสก่อน ซึ่งเป็นสิ่งที่เราควรทำเสมอ หลังจากนั้นจะไล่วาดจุดที่ถูกบันทึกไว้ทีละจุด จุดที่วาดจะวาดเป็นวงกลมด้วย `fillOval` โดยกำหนดสีตาม `inkLevel` ที่บันทึกไว้ และกำหนดขนาดของเส้นสัมพันธ์กับ `inkLevel` ด้วย (ยิ่งลากยาวเส้นยิ่งบางลง)

สังเกตว่าอินเทอร์เฟซ `MouseListener` และ `MouseMotionListener` มีเมทอดรวมกันถึง 7 เมทอด แต่เราใช้จริงเพียง 3 เมทอด ด้วยความที่เป็นอินเทอร์เฟซ ถึงเราจะใช้เพียง 3 เมทอด เราก็ต้องโอเวอร์ไรด์หมดทั้ง 7 เมทอด โดยปล่อยให้เมทอดอื่น ๆ ที่ไม่ใช้เป็นเมทอดว่าง ๆ ไป ซึ่งเป็นกระบวนการที่ไม่สะดวกนัก

ในกรณีของการจัดการเมาส์ เราจะมีคลาสพิเศษตัวหนึ่งชื่อว่า `MouseAdapter` ซึ่งจะอิมพลีเมนต์ทั้งอินเทอร์เฟซ `MouseListener` และ `MouseMotionListener` และโอเวอร์ไรด์เมทอดทั้งหมด โดยทิ้งเป็นเมทอดว่าง ๆ เอาไว้ เราสามารถเลือกที่จะซับคลาสจาก `MouseAdapter` แล้วโอเวอร์ไรด์เฉพาะเมทอดที่ต้องการใช้ แทนการอิมพลีเมนต์ `MouseListener` และ `MouseMotionListener` แล้วต้องโอเวอร์ไรด์ทุกเมทอดได้ คลาส `MouseAdapter` ถือเป็นคลาสอำนวยความสะดวกกลุ่มที่เรียกว่า adapter class ซึ่งทำให้เราไม่ต้องโอเวอร์ไรด์เมทอดที่เราไม่ได้ใช้ เราสามารถเลือกใช้ adapter class หรือจะอิมพลีเมนต์อินเทอร์เฟซโดยตรงก็ได้

ส่วนของ `main` เป็นดังนี้

````java
import javax.swing.JFrame;
import javax.swing.SwingUtilities;

public class DrawingBoardTest {
    public DrawingBoardTest() {
        DrawingBoard panel = new DrawingBoard();

        JFrame application = new JFrame();

        application.setDefaultCloseOperation(JFrame.EXIT_ON_CLOSE);

        application.add(panel);
        application.setSize(500, 500);
        application.setVisible(true);
    }

    public static void main(String[] args) {
        SwingUtilities.invokeLater(new Runnable() {
            @Override
            public void run() {
                new DrawingBoardTest();
            }
        });
    }
}
````

ในส่วนของ `main` มีส่วนที่แตกต่างจากปกติตรงที่เราใช้ `SwingUtilities.invokeLater()` และห่อโค้ดส่วนการสร้างหน้าโปรแกรมไว้ในคอนสตรักเตอร์ของ `DrawingBoardTest` และห่อการเรียกเอาไว้ใน `Runnable.run()` อีกที เราทำแบบนี้มาก่อนหน้านี้แล้วครั้งหนึ่งในบทก่อนหน้า การใช้ `invokeLater()` เป็นรูปแบบที่แนะนำโดย Sun Microsystems และ Oracle เนื่องจากข้อกำหนดว่าเราต้องไม่ทำกระบวนการที่เกี่ยวข้องกับ GUI ในเทรดอื่นนอกจาก EDT ซึ่งการเรียก `invokeLater()` จะเป็นการส่งงานไปให้ EDT เป็นผู้กระทำ

## การเคลื่อนที่

เราจะเริ่มต้นทำเกมอย่างง่ายกัน โปรแกรมตัวอย่างนี้จะโหลดรูปดาวขึ้นมา และรับอีเวนต์จากแป้นพิมพ์เพื่อขยับดาวไปทางซ้ายหรือขวาตามปุ่มลูกศร

````java
import java.awt.Graphics;
import java.awt.Color;
import java.awt.event.KeyEvent;
import java.awt.event.KeyListener;
import java.awt.image.BufferedImage;
import java.io.File;
import java.io.IOException;
import javax.imageio.ImageIO;
import javax.swing.JPanel;

public class GameScene extends JPanel {
    private BufferedImage star;
    private int positionX = 180;
    private int positionY = 120;

    public GameScene() {
        setBackground(Color.BLACK);
        setFocusable(true);

        // Loading star image
        try {
            star = ImageIO.read(new File("star.png"));
            if (star == null) {
                System.err.println("Unrecognized image type.");
                System.exit(1);
            }
        } catch (IOException e) {
            System.err.println("Error while loading image.");
            System.exit(1);
        }

        addKeyListener(new KeyListener() {
            @Override
            public void keyTyped(KeyEvent e) {
                // Do nothing
            }

            @Override
            public void keyPressed(KeyEvent e) {
                switch (e.getKeyCode()) {
                    case KeyEvent.VK_LEFT:
                        positionX -= 10;
                        repaint();
                        break;
                    case KeyEvent.VK_RIGHT:
                        positionX += 10;
                        repaint();
                        break;
                }
            }

            @Override
            public void keyReleased(KeyEvent e) {
                // Do nothing
            }
        });
    }

    @Override
    public void paintComponent(Graphics g) {
        super.paintComponent(g);

        g.drawImage(star, positionX, positionY, null);
    }
}
````

คลาส `GameScene` เป็นหน้าหลักของเกมและเป็นซับคลาสของ `JPanel` ในคอนสตรักเตอร์เราจะกำหนดพื้นหลังให้เป็นสีดำ และตั้งค่า `setFocusable(true)` ซึ่งเป็นขั้นตอนที่สำคัญมาก เราเรียกคอมโพเนนต์ที่ถูกเลือกใช้งานอยู่ว่ากำลังถูกโฟกัส (focus) ซึ่งคอมโพเนนต์ที่ได้รับโฟกัสจะได้รับอีเวนต์จากแป้นพิมพ์ ถ้าเราไม่กำหนดให้ `GameScene` ของเรารับโฟกัสได้ เราจะไม่สามารถรับอีเวนต์การกดปุ่มต่าง ๆ บนแป้นพิมพ์ได้เลย และทำให้เกมของเราไม่สามารถตอบสนองผู้เล่นได้

คอนสตรักเตอร์ของ `GameScene` ยังโหลดรูปและกำหนดตัวจัดการอีเวนต์แป้นพิมพ์ด้วย ซึ่งเราจะพูดถึงในรายละเอียดอีกครั้ง นอกจากนั้นก็มีเมทอด `paintComponent()` ที่เราโอเวอร์ไรด์เพื่อให้วาดรูปดาว ณ ตำแหน่งที่เรากำหนด ซึ่งเราจะให้เปลี่ยนไปตามปุ่มซ้ายขวาบนแป้นพิมพ์ ในส่วนของเมทอดนี้ อย่าลืมเรียก `paintComponent()` ของซูเปอร์คลาสด้วย ไม่อย่างนั้นการลงสีพื้นหลังของเราอาจจะผิดปกติได้

````java
import javax.swing.SwingUtilities;
import javax.swing.JFrame;

public class BasicGame {
    public static void main(String[] args) {
        SwingUtilities.invokeLater(new Runnable() {
            @Override
            public void run() {
                JFrame game = new JFrame();
                game.setDefaultCloseOperation(JFrame.EXIT_ON_CLOSE);
                game.setSize(400, 300);
                game.add(new GameScene());
                game.setVisible(true);
            }
        });
    }
}
````

ส่วนของ `main()` อยู่ในคลาส `BasicGame` ซึ่งเราจะยึดรูปแบบการใช้ `invokeLater()` เพื่อให้ไปรันโค้ดที่เกี่ยวข้องกับกราฟิกส์หรือ GUI ใน EDT ตามข้อแนะนำของ Sun/Oracle แทนที่จะรันใน main thread โดยตรง

### การโหลดรูป

ส่วนของการโหลดรูปเป็นดังนี้

```java
try {
    star = ImageIO.read(new File("star.png"));
    if (star == null) {
        System.err.println("Unrecognized image type.");
        System.exit(1);
    }
} catch (IOException e) {
    System.err.println("Error while loading image.");
    System.exit(1);
}
```

เราใช้ `BufferedImage` เป็นคลาสสำหรับแทนรูปที่โหลดเข้ามา โดยใช้ `ImageIO.read()` ในการโหลดจากไฟล์รูปภาพซึ่งกำหนดโดย `new File("star.png")`

ผลของการโหลดรูปภาพอาจเป็น `null` ได้ถ้า Java ไม่รู้จักชนิดของรูปภาพนั้นและไม่สามารถโหลดได้ และถ้าไฟล์รูปนั้นไม่มีอยู่ หรือเกิดปัญหาในการเปิดไฟล์รูปขึ้นมา `ImageIO.read()` ก็สามารถที่จะแจ้งเป็น `IOException` ได้ ซึ่งทั้ง 2 กรณีเราจัดการโดยการแสดงข้อความผิดพลาดและจบการทำงานของโปรแกรม

### การจัดการแป้นพิมพ์

การจัดการแป้นพิมพ์จะใช้อินเทอร์เฟซ `KeyListener` ซึ่งมีเมทอด 3 เมทอด ได้แก่ `keyPressed()` ซึ่งจะถูกเรียกเมื่อมีการกดปุ่มใด ๆ บนแป้นพิมพ์ `keyReleased()` ซึ่งจะถูกเรียกเมื่อมีการปล่อยปุ่มใด ๆ บนแป้นพิมพ์ และ `keyTyped()` ซึ่งจะถูกเรียกเมื่อมีการปล่อยปุ่มที่เป็นตัวอักษรหรือสัญลักษณ์ต่าง ๆ บนแป้นพิมพ์ ปุ่มพิเศษ เช่น Control, Shift, Alt, F1 จะไม่ทำให้เกิดการเรียก `keyTyped()`

```java
addKeyListener(new KeyListener() {
    @Override
    public void keyTyped(KeyEvent e) {
        // Do nothing
    }

    @Override
    public void keyPressed(KeyEvent e) {
        switch (e.getKeyCode()) {
            case KeyEvent.VK_LEFT:
                positionX -= 10;
                repaint();
                break;
            case KeyEvent.VK_RIGHT:
                positionX += 10;
                repaint();
                break;
        }
    }

    @Override
    public void keyReleased(KeyEvent e) {
        // Do nothing
    }
});
```

เราต้องโอเวอร์ไรด์ทั้ง 3 เมทอด แต่ในที่นี้เราใช้เพียงเมทอดเดียวคือ `keyPressed()` โดยเราจัดการให้เมื่อมีการกดปุ่มซ้ายหรือขวา เราจะขยับตำแหน่งในแกน x ของรูปดาวและสั่ง `repaint()` ให้เกิดการวาดใหม่

สิ่งสำคัญที่ต้องระวังคือ เราต้องกำหนดให้พาเนลของเราสามารถรับโฟกัสได้ด้วย `setFocusable(true)` มิฉะนั้นตัวจัดการแป้นพิมพ์ของเราจะไม่ถูกเรียกเลย

ข้อสังเกตอย่างหนึ่งที่เกิดขึ้นกับตัวอย่างนี้คือ ดาวจะขยับเมื่อเรากดปุ่มเท่านั้น ถ้าเรากดค้างไว้ดาวจะไม่ขยับต่อ ต้องปล่อยปุ่มแล้วกดใหม่ ทั้งนี้เป็นเพราะอีเวนต์จะเกิดขึ้นเฉพาะตอนกดปุ่มและปล่อยปุ่มเท่านั้น ไม่เกิดอีเวนต์ตอนเรากดปุ่มค้างเอาไว้ ดังนั้น ถ้าจะนำไปใช้กับเกมซึ่งให้กดปุ่มค้างได้ เราต้องมีการจัดการเพิ่มเติม ซึ่งเราจะทำในตัวอย่างเกมต่อไปนี้

## เกม

เกมมีหลักการทำงานพื้นฐานคือการประมวลผลเป็นเฟรม เฟรมคือภาพนิ่งแต่ละภาพที่แสดงต่อเนื่องกันจนดูเหมือนภาพเคลื่อนไหว ในแต่ละเฟรม เกมจะต้องทำกระบวนการต่อไปนี้

1. ตรวจสอบอินพุต
2. อัปเดตสถานะของวัตถุต่าง ๆ ภายในเกมและข้อมูลของตัวเกมเอง
3. วาดเฟรม ซึ่งประกอบด้วยฉาก วัตถุต่าง ๆ ในเกม และองค์ประกอบอื่น ๆ ของเกม เช่น คะแนน สถานะของตัวละคร

การทำงานทั้ง 3 อย่างนี้จะทำซ้ำ ๆ กันไปอย่างต่อเนื่องโดยมีระยะห่างของเวลาระหว่างเฟรมเท่า ๆ กัน การทำซ้ำนี้อาจจะเกิดขึ้นโดยการใช้ลูปวนซ้ำทั้ง 3 กระบวนการนี้ไปเรื่อย ๆ ที่เรียกว่า game loop หรืออาจจะใช้ตัวนับเวลาควบคุมการเริ่มทำงานของแต่ละเฟรม

จำนวนเฟรมที่มีการวาดต่อหนึ่งหน่วยเวลามีผลอย่างมากต่อความรู้สึกต่อเนื่องของผู้เล่น เราต้องการให้ได้จำนวนเฟรมมาก ๆ เพื่อให้เกมดูลื่นไหล แต่การประมวลผลที่เกิดขึ้นต่อเฟรมก็เป็นข้อจำกัดที่ทำให้เราไม่สามารถทำให้มีความถี่เฟรมได้เท่าที่เราต้องการ

การควบคุมจำนวนเฟรมต่อหน่วยเวลามักจะมี 2 รูปแบบหลัก แบบแรกจะพยายามควบคุมให้มีจำนวนเฟรมต่อวินาทีที่แน่นอน ซึ่งมีข้อดีตรงที่การคำนวณการเคลื่อนที่ของวัตถุในเกมจะง่าย เพราะเวลาต่อเฟรมค่อนข้างแน่นอน แต่จะขาดความยืดหยุ่นและไม่สามารถใช้ประโยชน์จากเครื่องคอมพิวเตอร์ที่มีสมรรถนะสูงกว่าได้ ส่วนแบบที่สองจะใช้เป็นลูป เมื่อประมวลผลเฟรมหนึ่งเสร็จก็จะประมวลผลเฟรมต่อไปทันที แบบนี้จำนวนเฟรมต่อวินาทีจะไม่คงที่ อาจจะเปลี่ยนไปตามความซับซ้อนของเกมในขณะนั้น ในรูปแบบนี้ การคำนวณการเคลื่อนที่ในเกมจะอิงกับเวลาที่ผ่านไปนับจากเฟรมก่อนหน้า ซึ่งเรียกว่า delta time

ตัวอย่างที่เราจะดูต่อไปนี้จะเป็นแบบจำนวนเฟรมต่อวินาทีคงที่และใช้ตัวนับเวลาควบคุมการเริ่มต้นของแต่ละเฟรม

เรามีคลาสที่สำคัญ 2 คลาส ได้แก่คลาส `GameScene` ซึ่งควบคุมการดำเนินไปของฉากหนึ่งฉากในเกม โดยจะควบคุมการทำงานเป็นเฟรมและคอยอัปเดตและวาดวัตถุในเกมในแต่ละเฟรม นอกจากนี้ยังทำหน้าที่จัดการกับอินพุตจากแป้นพิมพ์ด้วย ส่วนอีกคลาสคือ `GameObject` ซึ่งเป็นคลาสแทนอ็อบเจกต์ต่าง ๆ ที่จะปรากฏในเกมได้ เช่น ตัวละครของเรา ศัตรู และกระสุน

````java
import java.awt.Graphics;
import java.awt.Color;
import java.awt.event.ActionEvent;
import java.awt.event.ActionListener;
import java.awt.event.KeyEvent;
import java.awt.event.KeyAdapter;
import javax.swing.JPanel;
import javax.swing.Timer;
import java.util.ArrayList;
import java.util.Iterator;

public abstract class GameScene extends JPanel {
    public static final int KEY_UP = 0;
    public static final int KEY_DOWN = 1;
    public static final int KEY_LEFT = 2;
    public static final int KEY_RIGHT = 3;
    public static final int KEY_FIRE = 4;

    public enum KeyState { UP, DOWN }

    private KeyState[] keyStates = new KeyState[5];

    private static final int DEFAULT_FRAME_RATE = 50;
    private long previousTime;

    private ArrayList<GameObject> gameObjects = new ArrayList<>();
    private ArrayList<GameObject> waitList = new ArrayList<>();

    public GameScene() {
        this(DEFAULT_FRAME_RATE);
    }

    public GameScene(int frameRate) {
        setBackground(Color.BLACK);
        setFocusable(true);
        setDoubleBuffered(true);

        addKeyListener(new KeyAdapter() {
            @Override
            public void keyPressed(KeyEvent e) {
                switch (e.getKeyCode()) {
                    case KeyEvent.VK_UP:
                        keyStates[KEY_UP] = KeyState.DOWN;
                        break;
                    case KeyEvent.VK_DOWN:
                        keyStates[KEY_DOWN] = KeyState.DOWN;
                        break;
                    case KeyEvent.VK_LEFT:
                        keyStates[KEY_LEFT] = KeyState.DOWN;
                        break;
                    case KeyEvent.VK_RIGHT:
                        keyStates[KEY_RIGHT] = KeyState.DOWN;
                        break;
                    case KeyEvent.VK_SPACE:
                        keyStates[KEY_FIRE] = KeyState.DOWN;
                        break;
                }
            }

            @Override
            public void keyReleased(KeyEvent e) {
                switch (e.getKeyCode()) {
                    case KeyEvent.VK_UP:
                        keyStates[KEY_UP] = KeyState.UP;
                        break;
                    case KeyEvent.VK_DOWN:
                        keyStates[KEY_DOWN] = KeyState.UP;
                        break;
                    case KeyEvent.VK_LEFT:
                        keyStates[KEY_LEFT] = KeyState.UP;
                        break;
                    case KeyEvent.VK_RIGHT:
                        keyStates[KEY_RIGHT] = KeyState.UP;
                        break;
                    case KeyEvent.VK_SPACE:
                        keyStates[KEY_FIRE] = KeyState.UP;
                        break;
                }
            }
        });

        previousTime = System.currentTimeMillis();

        Timer timer = new Timer(1000/frameRate, new ActionListener() {
            @Override
            public void actionPerformed(ActionEvent e) {
                updateAll();
            }
        });

        timer.start();
    }

    public void addGameObject(GameObject object) {
        waitList.add(object);
    }

    public KeyState getKeyState(int key) {
        return keyStates[key];
    }

    private void updateAll() {
        long currentTime = System.currentTimeMillis();
        double deltaTime = (currentTime - previousTime) / 1000.0f;
        previousTime = currentTime;

        // Call update() on game scene and all game objects
        update(deltaTime);
        for (GameObject gameObject : gameObjects) {
            gameObject.update(deltaTime);
        }

        // Prune dead game objects
        Iterator<GameObject> it = gameObjects.iterator();
        while (it.hasNext()) {
            GameObject gameObject = it.next();
            if (gameObject.isDestroyed()) {
                it.remove();
            }
        }

        // Add newly queued game objects
        for (GameObject object : waitList) {
            gameObjects.add(object);
        }
        waitList.clear();

        repaint();
    }

    @Override
    public void paintComponent(Graphics g) {
        super.paintComponent(g);
        render(g);
        for (GameObject gameObject : gameObjects) {
            gameObject.draw(g);
        }
    }

    public abstract void update(double deltaTime);
    public abstract void render(Graphics g);
}
````

คลาส `GameScene` ประกอบด้วยคอนสตรักเตอร์ซึ่งทำหน้าที่ตั้งค่าที่จำเป็นในตอนเริ่มต้น ได้แก่ กำหนดให้พาเนลรับโฟกัสได้ซึ่งจำเป็นสำหรับการรับอินพุตจากแป้นพิมพ์ เปิดการใช้งาน double buffering (`setDoubleBuffered(true)`) ซึ่งทำให้พาเนลมีการวาดโดยใช้บัฟเฟอร์คู่ เป็นบัฟเฟอร์ที่กำลังแสดงผล และบัฟเฟอร์ที่เรากำลังวาดรูปใหม่แยกกัน เมื่อวาดเสร็จก็จะสลับบัฟเฟอร์วาดมาเป็นบัฟเฟอร์แสดงผล และสลับบัฟเฟอร์แสดงผลไปเป็นบัฟเฟอร์วาด การทำบัฟเฟอร์คู่ทำให้การวาดไม่เกิดอาการกะพริบ

อีกส่วนหนึ่งของคอนสตรักเตอร์จะกำหนดตัวจัดการแป้นพิมพ์โดยใช้ `KeyAdapter` ซึ่งต่างจาก `KeyListener` ตรงที่เราไม่ต้องโอเวอร์ไรด์ทุกเมทอดก็ได้ และในตัวจัดการแป้นพิมพ์นี้เราจะเก็บสถานะแต่ละปุ่มด้วยว่าถูกกดอยู่หรือถูกปล่อยอยู่ ตัวจัดการแป้นพิมพ์นี้จะไม่เข้าไปควบคุมตัวเกมโดยตรง แต่จะทำหน้าที่เพียงคอยอัปเดตสถานะของแต่ละปุ่มเท่านั้น ขณะที่เกมกำลังดำเนินไปในแต่ละเฟรม จะเป็นหน้าที่ของวัตถุในเกมเองที่จะต้องตรวจสอบสถานะของแต่ละปุ่มและตอบสนองตามความเหมาะสม

ส่วนสุดท้ายของคอนสตรักเตอร์เป็นการตั้งเวลาให้กับ `timer` โดยดูจากจำนวนเฟรมต่อวินาทีที่ต้องการ `timer` จะเรียกเมทอด `updateAll()` ทุก ๆ ช่วงเวลาตามที่ตั้ง

เมทอด `updateAll()` จะทำหน้าที่สั่งให้วัตถุในเกมทุกตัวและตัว `GameScene` เองอัปเดตตัวเองโดยการเรียกเมทอด `update()` ของทุก ๆ ตัวพร้อมส่ง `deltaTime` ไปให้เพื่อบอกว่าเวลาผ่านไปกี่วินาทีแล้วหลังจากเฟรมก่อนหน้า (ซึ่งจริง ๆ จะน้อยกว่าวินาที)

หลังจากที่สั่ง `update()` ทุกตัวแล้ว ก็จะตรวจสอบสถานะของวัตถุแต่ละตัวว่าถูกทำลายแล้วหรือยัง (เช่น ถูกยิงหรือชนกัน) และลบทิ้งจาก `gameObjects` ซึ่งเป็น `ArrayList` ที่เก็บ `GameObject` ทุกตัวที่อยู่ในฉาก เราใช้ `Iterator` เนื่องจากการลบวัตถุออกจาก `ArrayList` ภายในลูปที่จัดการกับตัว `ArrayList` นั้นเองไม่สามารถทำโดยใช้ลูป `for` ปกติได้

เมื่อมีการสร้าง `GameObject` ขึ้นมาใหม่โดยใครก็ตาม จะต้องมาบรรจุเข้าไปใน `GameScene` โดยการการเรียกเมทอด `addGameObject()` ซึ่งเมทอดนี้จะยังไม่บรรจุอ็อบเจกต์ที่ใส่เข้ามาลงไปใน `ArrayList` `gameObjects` โดยตรงทันที เพราะ Java ไม่อนุญาตให้เกิดการบรรจุข้อมูลใหม่ขณะที่กำลังวนลูปจัดการกับ `ArrayList` ตัวนั้นโดยตรงอยู่ เราจึงบรรจุเข้าไปใน `waitList` ก่อน

`updateAll()` จะมีส่วนสุดท้ายที่นำอ็อบเจกต์ใน `waitList` มาบรรจุเข้าไปใน `gameObjects` โดยใช้การลูปบน `waitList` เราจำเป็นต้องทำแบบนี้เพราะเราไม่สามารถ `add()` เข้าไปใน `gameObjects` ขณะที่กำลังวนลูปบน `gameObjects` เองได้

เมื่อสิ้นสุดการอัปเดตข้อมูลทุกอย่างแล้ว `updateAll()` จะเรียก `repaint()` เพื่อร้องขอให้เรียก `paintComponent()`

`paintComponent()` จะถูกเรียกตามมา โดยจะไปเรียกเมทอด `render()` ของ `GameScene` เพื่อให้ `GameScene` วาดในส่วนของตัวเอง และเรียก `draw()` ของ `GameObject` ทุกตัวที่อยู่ในฉากให้วาดตัวเอง

นอกจากนี้ยังมีเมทอด `getKeyState()` ที่เอาไว้ตรวจสอบสถานะของปุ่มแต่ละปุ่มว่ากดหรือปล่อยอยู่

สุดท้ายคือเมทอด `update()` และ `render()` ของ `GameScene` ซึ่งเป็นเมทอดนามธรรม (abstract method) ซึ่งหมายความว่า `GameScene` เองก็เป็นคลาสนามธรรม ทำหน้าที่จัดการเฟรมและแป้นพิมพ์ แต่การวาดและการจัดการสถานะจริงให้เป็นหน้าที่ของโค้ดของซับคลาส

````java
import java.awt.Graphics;
import java.awt.Rectangle;

public abstract class GameObject {
    private boolean destroyed = false;

    public GameObject() {
        destroyed = false;
    }

    public abstract void update(double deltaTime);

    public abstract void draw(Graphics g);

    public abstract Rectangle getBounds();

    public boolean collides(GameObject other) {
        if (isDestroyed() || other.isDestroyed()) {
            return false;
        } else {
            return getBounds().intersects(other.getBounds());
        }
    }

    public void destroy() {
        destroyed = true;
    }

    public boolean isDestroyed() {
        return destroyed;
    }
}
````

`GameObject` เป็นคลาสนามธรรมที่แทนวัตถุในเกมทุกชนิด ประกอบด้วยเมทอด `update()` สำหรับอัปเดตสถานะของวัตถุ และ `draw()` สำหรับวาดตัววัตถุ ทั้งสองเมทอดนี้จะถูกเรียกให้ทำงานทุกเฟรม โดย `update()` จะถูกเรียกก่อน และ `draw()` จะถูกเรียกตาม เราจะต้องโอเวอร์ไรด์เมทอดเหล่านี้เพื่อกำหนดพฤติกรรมและการวาดตามที่เราต้องการ

วัตถุจะมีสถานะว่าเป็นปกติหรือถูกทำลายแล้ว เราจะเรียกเมทอด `destroy()` เพื่อกำหนดให้วัตถุนี้มีสถานะเป็นถูกทำลายแล้ว

เราจะโอเวอร์ไรด์ `getBounds()` เพื่อให้คืนค่าเป็นกรอบสี่เหลี่ยมที่แทนขอบเขตของวัตถุ เราใช้กรอบสี่เหลี่ยมนี้ในการทดสอบว่าวัตถุเกิดการชนกับวัตถุอื่นหรือไม่ โดยมีเมทอด `collides()` สำหรับตรวจสอบการชน วัตถุที่ถูกทำลายแล้วจะไม่เกิดการชน

ทั้ง `GameScene` และ `GameObject` ต่างก็เป็นคลาสนามธรรม การทำงานของเกมจริง ๆ จะเกิดขึ้นจากซับคลาสของทั้งสองคลาสนี้ ซึ่งได้แก่คลาสต่อไปนี้

````java
import java.awt.Graphics;
import java.awt.Color;
import java.util.ArrayList;
import java.util.Random;

public class ShootingScene extends GameScene {
     private SpaceShip player;
     private ArrayList<GameObject> enemies = new ArrayList<GameObject>();
     private ArrayList<GameObject> bullets = new ArrayList<GameObject>();

     private Random random;
     private int score;
     private int respawnCount = 3;

     public ShootingScene() {
         player = new SpaceShip(this, 400, 500);
         addGameObject(player);
         random = new Random();
     }

     @Override
     public void update(double deltaTime) {
         for (GameObject bullet : bullets) {
             for (GameObject enemy : enemies) {
                 if (bullet.collides(enemy)) {
                     bullet.destroy();
                     enemy.destroy();

                     addScore(10);
                     respawnCount++;
                 }
             }
         }

         while (respawnCount > 0) {
             Alien alien = new Alien(this, 100 + random.nextInt(700), 100 + random.nextInt(200));
             enemies.add(alien);
             addGameObject(alien);
             respawnCount--;
         }
     }

     @Override
     public void render(Graphics g) {
         g.setColor(Color.WHITE);
         g.drawString("Score: " + score, 10, 20);
     }

     public void fire(Bullet bullet) {
         bullets.add(bullet);
         addGameObject(bullet);
     }

     public void addScore(int points) {
         score += points;
     }
}
````

คลาส `ShootingScene` สืบทอดจาก `GameScene` โดยทำหน้าที่สร้างยานผู้เล่น (`Player`) และยานศัตรู (`Alien`) โดยจะโอเวอร์ไรด์เมทอด `update()` และ `render()`

เมทอด `update()` จะถูกเรียกก่อน `render()` ในทุกเฟรม และทำหน้าที่ตรวจสอบว่ากระสุน (`Bullet`) ชนกับศัตรูตัวไหนบ้างหรือไม่ ถ้าชนก็จะสั่งทำลายทั้งกระสุนและศัตรูโดยการเรียกเมทอด `destroy()` และเพิ่มคะแนนให้ เราจะนับจำนวนศัตรูที่ถูกทำลายไปเพื่อเอาไปสร้างขึ้นมาใหม่ด้วยตัวแปร `respawnCount` และในเมทอดนี้ในตอนหลังก็จะสร้างยานศัตรูขึ้นใหม่ตาม `respawnCount` สังเกตว่าเรากำหนดจำนวนศัตรูตั้งต้นไว้ในตอนแรกด้วย `respawnCount` ด้วย ซึ่งยานศัตรูตอนแรกเริ่มก็จะถูกสร้างใน `update()` นี่เอง

การสร้างยานศัตรูแต่ละครั้งเราจะบรรจุยานศัตรูลงไปทั้งในรายการของ `GameScene` ด้วย `addGameObject` และบรรจุลงในรายการ `enemies` ต่างหากด้วยเพื่อให้การจัดการแยกตามประเภทของวัตถุสะดวกขึ้น (เราแยก `Player`, `Alien` และ `Bullet` ออกจากกัน)

เมทอด `render()` ทำเพียงการวาดคะแนนลงบนจอภาพ

นอกจากนี้ยังมีเมทอด `fire()` ซึ่งจะถูกเรียกเวลาที่ตัวผู้เล่นมีการปล่อยกระสุน โดยจะบรรจุกระสุนลงทั้งในรายการของ `GameScene` และ `bullets` เช่นกัน เราจำเป็นต้องบรรจุทุกวัตถุลงใน `GameScene` เสมอ มิฉะนั้นวัตถุนั้นจะทำงานไม่ได้

````java
import java.awt.*;

public class Alien extends GameObject {
    private GameScene scene;
    private int x, y;
    private int speedX = 10;
    private int speedY = 10;

    public Alien(GameScene scene, int x, int y) {
        this.scene = scene;
        this.x = x;
        this.y = y;
    }

    @Override
    public void update(double deltaTime) {
        if (x < 100) {
            speedX = 10;
        } else if (x > 700) {
            speedX = -10;
        }
        if (y < 100) {
            speedY = 10;
        } else if (y > 300) {
            speedY = -10;
        }
        x += speedX;
        y += speedY;
    }

    @Override
    public void draw(Graphics g) {
        g.setColor(Color.BLUE);
        g.fillRect(x, y, 50, 50);
    }

    @Override
    public Rectangle getBounds() {
        return new Rectangle(x, y, 50, 50);
    }
}
````

คลาส `Alien` เป็นยานของศัตรู เราโอเวอร์ไรด์เมทอด `update()` เพื่อกำหนดการเคลื่อนที่เป็นแนวทแยง โดยให้สะท้อนกลับถ้าชนขอบ เราวาดยานศัตรูเป็นสี่เหลี่ยมธรรมดา

````java
import java.awt.*;
import java.io.File;
import java.io.IOException;
import javax.imageio.ImageIO;

public class SpaceShip extends GameObject {
    private ShootingScene scene;
    private Image image;
    private int x, y;
    private int speed = 10;

    private static final double FIRING_INTERVAL = 0.5;
    private double lastFiringElapsed = 0.0;

    public SpaceShip(ShootingScene scene, int x, int y) {
        this.scene = scene;

        // Load player image
        try {
            image = ImageIO.read(new File("images/star.png"));
            if (image == null) {
                System.err.println("Unrecognized image type.");
                System.exit(1);
            }
        } catch (IOException e) {
            System.err.println("Error while loading image.");
            System.exit(1);
        }

        this.x = x - image.getWidth(null) / 2;
        this.y = y - image.getHeight(null) / 2;
    }

    @Override
    public void update(double deltaTime) {
        if (scene.getKeyState(GameScene.KEY_UP) == GameScene.KeyState.DOWN) {
            y -= speed;
        }
        if (scene.getKeyState(GameScene.KEY_DOWN) == GameScene.KeyState.DOWN) {
            y += speed;
        }
        if (scene.getKeyState(GameScene.KEY_LEFT) == GameScene.KeyState.DOWN) {
            x -= speed;
        }
        if (scene.getKeyState(GameScene.KEY_RIGHT) == GameScene.KeyState.DOWN) {
            x += speed;
        }
        if (scene.getKeyState(GameScene.KEY_FIRE) == GameScene.KeyState.DOWN) {
            if (lastFiringElapsed >= FIRING_INTERVAL) {
                scene.fire(new Bullet(scene, x, y));
                lastFiringElapsed = 0.0;
            }
        }

        lastFiringElapsed += deltaTime;
    }

    @Override
    public void draw(Graphics g) {
        g.drawImage(image, x, y, scene);
    }

    @Override
    public Rectangle getBounds() {
        return new Rectangle(x, y, image.getWidth(null), image.getHeight(null));
    }

    public int getX() {
        return x;
    }

    public int getY() {
        return y;
    }
}
````

คลาส `SpaceShip` เป็นยานของผู้เล่น คอนสตรักเตอร์จะทำหน้าที่โหลดรูปภาพของยานขึ้นมา ในส่วนเมทอด `update()` ซึ่งเราโอเวอร์ไรด์ เราจะมีการตรวจสอบสถานะของปุ่มควบคุมทิศทางแต่ละปุ่ม และปรับตำแหน่งของยานตามการควบคุม แต่ถ้าเป็นการกดปุ่มยิง (space bar) เราจะสร้างกระสุนจากคลาส `Bullet` แล้วบรรจุเข้าไปใน `ShootingScene` แต่การยิงจะมีการตรวจสอบก่อนด้วยว่าการยิงก่อนหน้านั้นห่างกันเกินเวลาที่กำหนดหรือไม่ ถ้าไม่เกินก็จะยังไม่สร้างกระสุน เพื่อให้ไม่เกิดการยิงรัวติดกันถี่เกินไป เรานับเวลาจากค่า `deltaTime` ที่ส่งเข้ามาในแต่ละรอบเฟรม

````java
import java.awt.*;

public class Bullet extends GameObject {
    private GameScene scene;
    private int x, y;
    private int speed = 20;

    public Bullet(GameScene scene, int x, int y) {
        this.scene = scene;
        this.x = x;
        this.y = y;
    }

    @Override
    public void update(double deltaTime) {
        y -= speed;
        if (y < 0) {
            destroy();
        }
    }

    @Override
    public void draw(Graphics g) {
        g.setColor(Color.WHITE);
        g.fillRect(x, y, 5, 10);
    }

    @Override
    public Rectangle getBounds() {
        return new Rectangle(x, y, 5, 10);
    }
}
````

คลาส `Bullet` แทนกระสุน ซึ่งจะเคลื่อนที่ทิศทางเดียวคือขึ้นไปด้านบน ในที่นี้เราวาดกระสุนด้วยสี่เหลี่ยม ถ้ากระสุนพ้นขอบจอด้านบนเราจะให้กระสุนสั่งทำลายตัวเอง

````java
import javax.swing.SwingUtilities;
import javax.swing.JFrame;

public class SpaceGame {
    public static void main(String[] args) {
        SwingUtilities.invokeLater(new Runnable() {
            @Override
            public void run() {
                JFrame game = new JFrame();
                game.setDefaultCloseOperation(JFrame.EXIT_ON_CLOSE);
                game.setSize(800, 600);
                game.add(new ShootingScene());
                game.setVisible(true);
            }
        });
    }
}
````

คลาส `SpaceGame` เป็นตัวขับเคลื่อนเกมนี้โดยการสร้างอ็อบเจกต์ `ShootingScene` และเช่นเดียวกับตัวอย่างที่แล้ว เราจะยึดรูปแบบการใช้ invokeLater() เพื่อให้ไปรันโค้ดที่เกี่ยวข้องกับกราฟิกส์หรือ GUI ใน EDT ตามข้อแนะนำของ Sun/Oracle แทนที่จะรันใน main thread โดยตรง
